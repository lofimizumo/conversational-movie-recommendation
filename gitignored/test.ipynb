```python

def getResponse(request):
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
                {"role": "system", "content": "You are a assistant, output information without any answering or conversation or explanation."},
                {"role": "user", "content": "from now on based on movie name provided,make csv as the only output, don't output conversation, columns separated using @ symbol, columns are: movie name,plot and question,plot column is the movie plot in around 150 words, question column looks like: can you recommend me a movie about... this question should be related to the movie plot"},
                {"role": "assistant", "content": "Understood, starting from now, I will provide the requested output in CSV format only without any extra comments or conversation. The columns are separated using the @ symbol, and the columns are Movie Name, Plot, and Question. The Plot column is the movie plot of the requested movie in around 150 words, and the Question column is a question related to the movie plot.Please provide me with the name of the movie you would like information about, and I will generate the CSV output for you."},
                {"role": "user", "content": request}
    ]
    )
    return response
```

```python
def format_result(txt):
    pattern = r'(?s)([\w\s\'",!?-]+)\@([\w\s\'",;.!?-]+)\@([\w\s\'",;.!?-]+)'
    matches = re.findall(pattern, txt)
    csv_string = "\n".join(["@".join(match) for match in matches])
    return csv_string
```
